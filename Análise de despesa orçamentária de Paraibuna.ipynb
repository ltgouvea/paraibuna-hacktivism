```python
import numpy as np;
import pandas as pd;

my_data = np.genfromtxt('./paraibuna/despesa_orcamentaria_maio18.csv', delimiter=';')

do = pd.read_csv('./paraibuna/despesa_orcamentaria_maio18.csv', ';');

do['Valor'] = do['Valor'].str.strip()

do['Valor'] = do['Valor'].str.replace(".", "").str.replace(",", ".").astype(float)

```

```python
fornecedores = do.groupby(['Fornecedor'])['Valor'].agg(
  lambda x : x.sum()
).sort_values(axis=0, ascending=False)

print(fornecedores);
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

ax = fornecedores.head(5).plot(kind='bar', title ="Gasto com Fornecedores", figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("Fornecedores", fontsize=12)
ax.set_ylabel("Valor em Reais", fontsize=12)
plt.show()
```
