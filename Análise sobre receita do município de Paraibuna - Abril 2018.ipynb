```python
import numpy as np;
import pandas as pd;

do = pd.read_csv('./datasets/receita_abril_2018.csv', ',')

do['Descricao'] = do['Descricao'].str.strip()

# do['No Mes'] = do['No Mes'].astype(float)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

receitas = do.groupby(['Descricao'])['No Mes'].agg(
  lambda x : x.sum()
).sort_values(axis=0, ascending=False)

fig = plt.figure(figsize=[8, 8])
ax = fig.add_subplot(111)

receitas = receitas.head(20)

labels = receitas.keys()
sizes = receitas

def absolute_value(val):
    a  = np.round(val/100.*sizes.sum(), 0)
    return a


patches, texts, autopct = ax.pie(sizes, autopct=absolute_value, shadow=False, startangle=140)
# plt.legend(patches, labels, loc="upper right")
ax.legend(patches, labels, loc="center left" ,
          bbox_to_anchor=(1, 0, 0.5, 1))
ax.set_title("Despesa Orçamentária de acordo com Programas municipais")
ax.axis('equal')
```
